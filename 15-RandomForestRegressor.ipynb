### Dependent - Independent Features

```python
X = df.drop("number_people", axis=1)
y = df["number_people"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
#Normalde ağaç bazlı olduğu için scale etmeye gerek yok. Ama yapmak isterseniz yapabilirsiniz.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
#bize y_test ve y_predict'i verecek iki metrikli fonksiyon oluşturalım
def calculate_model_metrics(true, predicted):
    mae = mean_absolute_error(true,predicted)
    mse = mean_squared_error(true,predicted)
    rmse = np.sqrt(mean_squared_error(true,predicted))
    r2 = r2_score(true,predicted)
    return mae, mse, rmse, r2
```

```python
#Hangi modelleri kullanacağımızı belirtelim (Dictionarydir Key: Value yani)
models = {
    "Linear Regression" : LinearRegression(),
    "Ridge": Ridge(),
    "Lasso" : Lasso(),
    "K-Neighbors Regression" : KNeighborsRegressor(),
    "Decision Tree Regression" : DecisionTreeRegressor(),
    "Random Forest Regression": RandomForestRegressor()
}
```

```python
#bunları direkt bir for loop'a alacağız
for i in range(len(list(models))):
    model = list(models.values())[i] #dictionary olduğu için modelimizi aldık ve kaçıncı indeksindeysek getirdik
    model.fit(X_train_scaled, y_train)
    
    y_train_pred  = model.predict(X_train_scaled) #train datasının da r2 score'unu falan alabiliriz bununla. overfittingi net görmemizi sağlar
    y_test_pred = model.predict(X_test_scaled)
    
    model_train_mae, model_train_mse, model_train_rmse, model_train_r2 = calculate_model_metrics(y_train, y_train_pred) #training accuracy
    model_test_mae, model_test_mse, model_test_rmse, model_test_r2 = calculate_model_metrics(y_test, y_test_pred) #test accuracy
                                  
    print(list(models.values())[i]) #model adımızı basar
    
    print("Evaluation for Training Set")
    print("RMSE:" , model_train_rmse)
    print("MSE:" , model_train_mse)
    print("MAE:" , model_train_mae)
    print("R2:" , model_train_r2)

    print("-------------------------------------------------")
    
    print("Evaluation for Test Set")
    print("RMSE:" , model_test_rmse)
    print("MSE:" , model_test_mse)
    print("MAE:" , model_test_mae)
    print("R2:" , model_test_r2)
  
    print("-------------------------------------------------")
    print("\n") #boşluk bırakıp devam edelim çünkü tüm modeller için yazdıracak
    
```

### Hyperparameter Tuning

```python
#KNN ve Random Forest'ı alalım.
knn_params  = {"n_neighbors" : [2,3,10,20,40,50]}
rf_params = {
    "max_depth" : [5,8,10,15,None],
    "max_features": ["sqrt","log2",5,7,10],
    "min_samples_split": [2,8,12,20],
    "n_estimators" : [100,200,500,1000]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomcv_models = [
    ("KNN", KNeighborsRegressor(), knn_params),
    ("RF", RandomForestRegressor(), rf_params)
                  
]
```

```python
for name, model, params in randomcv_models:
    randomcv = RandomizedSearchCV(estimator=model, param_distributions=params, n_iter=100, cv=3)
    randomcv.fit(X_train_scaled, y_train)
    print("Best params for: ", name, randomcv.best_params_)
```

```python
#Best params for:  KNN {'n_neighbors': 2}
#Best params for:  RF {'n_estimators': 500, 'min_samples_split': 2, 'max_features': 7, 'max_depth': None}
```
