```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("6-bank_customers.csv")
```

```python
df.columns
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Data scale edilmiş,encode edilmiş temiz bir data olduğu için direkt X,y ile başlayalım.
X = df.drop("subscribed",axis=1) #hedef değişkeni drop ederiz yani dependent feature'dır
y = df["subscribed"] #hedef olduğu için
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
logistic.fit(X_train,y_train)
y_pred = logistic.predict(X_test)
```

```python
y_pred #predictionımız binary değer oldu
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print("Score: ", score)
print(classification_report(y_pred,y_test))
print("Confusion matrix: \n",confusion_matrix(y_pred,y_test))
```

```python
# Hyperparameter Tuning
model = LogisticRegression()
```

```python
#Algoritmanın accuracy'sini değiştirebilecek önemli hiperparametreler:
penalty = ["l1","l2","elasticnet"]
c_values = [100,10,1,0.1,0.01] #negatif olamaz
solver = ["newton-cg","lbfgs","liblinear","sag","saga","newton-cholesky"] #default burda lbfgs
```

```python
#bu hiperparametreleri sözlük (dictionary) şeklinde tutmak daha iyidir.
#keys:values
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
params
```

### GridSearchCV

```python
# GridSearchCv: Modelin en iyi ayarlarını bulma, belirlenen parametrelerin farklı kombinasyonlarını deneyip, modelin hangi kombinasyonla
# en iyi sonucu verdiğini bulur.
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
cv = StratifiedKFold() #default 5
```

```python
grid = GridSearchCV(estimator = model, param_grid = params, cv = cv, scoring = "accuracy", n_jobs = -1) #Cross Validation yapmak istiyorsak içine StratifiedKFold'u 
#estimator ve param_grid'i vermeniz gerekiyor.
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_ #hangi hiperparametreye hangi değeri vermemiz gerektiğini gösterir (key:value)
```

```python
grid.best_score_ #accuracy skorunu verir
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_pred,y_test)
print("Score: ", score)
print(classification_report(y_pred,y_test))
print("Confusion matrix: \n",confusion_matrix(y_pred,y_test))
```

### Random Search CV

```python
#RandomizedSearchCV, makine öğrenmesinde modelin hiperparametrelerini optimize etmek için kullanılan bir yöntemdir.
# GridSearchCV tüm olası parametre kombinasyonlarını tek tek dener.
# RandomizedSearchCV ise, bu kombinasyonlardan rastgele seçilmiş belli sayıda kombinasyonu dener.
# RandomizedSearchCV, çok büyük parametre aralıklarında daha hızlı çalışır, genelde çok iyi sonuçlar verir.
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = LogisticRegression()
```

```python
randomcv = RandomizedSearchCV(estimator=model, param_distributions=params,cv=5, n_iter=10, scoring="accuracy")
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_params_
```

```python
randomcv.best_score_
```

```python
y_pred = randomcv.predict(X_test)
```

```python
score = accuracy_score(y_pred,y_test)
print("Score: ", score)
print(classification_report(y_pred,y_test))
print("Confusion matrix: \n",confusion_matrix(y_pred,y_test))
```
