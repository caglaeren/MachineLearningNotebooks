```python
from lazypredict.Supervised import LazyRegressor
from sklearn import datasets
from sklearn.utils import shuffle
import numpy as np
```

```python
diabetes = datasets.load_diabetes() #scikit-learn'ün içinde hazır datalar barındıran diabet datası
```

```python
X, y = shuffle(diabetes.data, diabetes.target, random_state=13) #X'leri dataya verir, y'leri target'a veriyor ve bunları verirken de karıştırır
```

```python
X = X.astype(np.float32)
```

```python
#test train split kullanmadan numpy'da kendisi bu şekilde yapmış
offset = int(X.shape[0] * 0.9) #%90'ı traine, %10'u teste vermiş

X_train, y_train = X[:offset], y[:offset]
X_test, y_test = X[offset:], y[offset:]
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)

#verbose=0 : Ekrana ayrıntılı eğitim bilgisi yazdırmaz,eğer 1 yapılırsa hangi modelin eğitildiğini sırayla yazar
#ignore_warnings=False: Model eğitimi sırasında çıkan uyarıları gizlemez, uyarılar ekranda görülür
#custom_metrics=None: Eğer kendi değerlendirme metriğini tanımlamak istersen özel bir fonk verebilirsin, burada None olduğu için
#varsayılan metrikler yani (R², RMSE, vs.) kullanılır.
```

```python
models, predictions = reg.fit(X_train, X_test, y_train, y_test) #Eğitim verisi ile modelleri eğitir, test verisinde modelleri değerlendirir ve sonrasında tablo döndürür
```
